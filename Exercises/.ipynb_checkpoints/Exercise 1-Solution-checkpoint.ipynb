Wie lassen sich die Stundenwerte in ein Datum und Uhrzeit umrechnen?

Das Praktische an der Library ist, dass sie automatisch Schaltjahre usw. berücksichtigt:

````python
s1 = datetime.datetime(year=2021,month=1,day=1)
s2 = datetime.datetime(year=2020,month=1,day=1)
print("Höchste Temperatur 2021:", s1+datetime.timedelta(hours=int(high[0])))
print("Höchste Temperatur 2020:", s2+datetime.timedelta(hours=int(high[0])))

````

### Tagesmittelwert der Außentemperatur
Ermitteln Sie den Tagesmittelwert der Außentemperatur und stellen Sie sie  der stündlichen Außentemperatur gegenüber

````python
# Ermittlung des Tagesmittels der Lufttemperatur
tagesmittel = []
for day in range(365):
    daily_avg = TA[day*24:(day+1)*24].mean() # TA[von : bis] in Stunden (Range von 0-364)
    tagesmittel.append(daily_avg)

len(tagesmittel)
````

````python
plt.plot(TA, linewidth=0.3, alpha=0.5)
plt.plot(range(0,8760,24),tagesmittel, color="darkblue")
plt.ylabel("Außentemperatur [°C]")
plt.xlabel("Stunden")
````

### Heizgradtage
Ermitteln Sie die Heizgradtage $HGT_{20/12}$ , dh. für eine Außentemperatur von 12°C und eine Zieltemperatur von 20°C gemäß folgender Definition:

    Summe der Temperaturdifferenzen zwischen einer bestimmten konstanten Raumtemperatur und
     dem Tagesmittel der Lufttemperatur, falls diese gleich oder unter einer angenommenen Heizgrenztemperatur liegt.
    Definition der Heizgradtagzahl laut ON H5056-1

````python
HGT = 0
for day in tagesmittel: # "für jeden eintrag - nenn ihn jeweils day - im container (hier eine liste) tagesmittel "
    if day < 12.0:
        HGT += 20 - day # same as HGT = HGT + 20 - day

HGT
````

### HGT Funktion
Schreiben Sie nun eine Funktion $HGT(l, z, g)$, die die Heizgradtage ermittelt und
als Inputs die Liste der Werte, die Zieltemperatur und den Grenzwert der
Außentemperatur entgegen nimmt:

Vergleichen Sie mit dieser Funktion die $HGT_{20/12}$ und die $HGT_{22/14}$ für den Klimadatensatz

````python
def HGT(l, z, g):
    summe = 0
    for day in l:
        if day < g:
            summe += z - day

    return summe

print("HGT22/14:", HGT(tagesmittel, 22, 14))
print("HGT20/12:", round(HGT(tagesmittel, 20, 12),2)) 
````

### Generalization
Schreiben Sie eine Funktion **monthly**, die die Mittlere monatliche Außentemperatur einer Zeitreihe wieder als Zeitreihe zurück gibt.
* Erstellen Sie ein datetime array mittels ```np.arange(von, bis, dtype='datetime64[h]') ```  für das Jahr 2021 https://numpy.org/doc/stable/reference/arrays.datetime.html
* verwenden Sie dieses array als Index für einen pandas DataFrame
* Verwenden Sie die Pandas .resample() Methode 

````python
zeitreihe = np.arange('2021-01-01', '2022-01-01', dtype='datetime64[h]')
zeitreihe
````

````python
import pandas as pd

TSD = pd.DataFrame(TA, columns=["TA"], index=zeitreihe)
Month= TSD.resample("M").mean()
Month.plot(kind="bar")
````

````python
def monthly(timeseries):
    zeitreihe = np.arange('2021-01-01', '2022-01-01', dtype='datetime64[h]')
    import pandas as pd
    TSD = pd.DataFrame(timeseries, index=zeitreihe)
    return TSD.resample("M").mean()

monthly(TA)
````

### Ermittle den Tagesmittelwert der Außentemperatur mit pandas.resample()


````python
import pandas as pd

TSD = pd.DataFrame(TA, columns=["TA"], index=zeitreihe)

Daily = TSD.resample("D").mean()
plt.plot(TSD.TA)
plt.plot(Daily.TA, color="darkblue")
````

### (challenge) Ermittle die MONATLICHEN Heizgradtage
* Erstellen Sie eine Monatliche Zeitreihe [1-12]
* Ermitteln Sie für jeden Monat die HGT mit der Funktion HGT() aus 1.1.9 


````python
HGT_mon = []
for m, series in Daily.TA.groupby(Daily.index.month):
    HGT_mon.append(round(HGT(series, 20,12),2))
    print(f"Monat {m:2.0f} - Monatl. Mitteltemperatur: {round(series.mean(),2):3.2f}°C - HGT: {HGT_mon[-1]}")
````

````python
sum(HGT_mon)
````
